### Заполнение пропусков

В столбце `tenure` около 9% пропусков. Для успешного обучения моделей машинного обучения нужно от них избавится.

Взглянем на распределение этой величины:

```python
# Размер картинки
plt.gcf().set_size_inches(10, 6) 

# Строим график
tenure_hist = sns.histplot(df, x='tenure', kde=True, hue='exited')

# Редактируем название осей и даем имя графику
_ = tenure_hist.set(title='Диаграмма времени удержания клиентов', xlabel="Время(в годах)", ylabel="Количество")
```

Чтобы не образовывать пики на графике, пропуски заполним пропорционально изначальному распределению.

```python
# Распределение в долях
tenure_dist = df.tenure.value_counts(normalize=True)
# Применяем это же распределение к количеству пропусков(чтобы получить такое же)
tenure_dist = tenure_dist.values * df.tenure.isna().sum()
# Смотрим результат
tenure_dist
```

```python
print('Сумма до округления =', tenure_dist.sum())
```

```python
# Округляем 
tenure_dist_rounded = tenure_dist.round()
# Потерю данных при округлении прибавляем к последнему классу
tenure_dist_rounded[-1] = tenure_dist_rounded[-1] + tenure_dist.sum() - tenure_dist_rounded.sum()
# Преобразовываем тип
tenure_dist_rounded = tenure_dist_rounded.astype('int')
print('Сумма после округления =', tenure_dist_rounded.sum())
```

```python
# Преобразуем в pd.Series
tenure_dist_pd_series = pd.Series(tenure_dist_rounded, index=df.tenure.value_counts().index)
# Меняем тип индекса на целый
tenure_dist_pd_series.index = tenure_dist_pd_series.index.astype('int')
# Сортируем
tenure_dist_pd_series = tenure_dist_pd_series.sort_index()
# Результат
tenure_dist_pd_series
```

```python
# Создаем список куда будем складывать наши значения
list_integers_to_replace = []
# Заполняем список по принципу index - значение, value - сколько раз положить значение(он же индекс) в список
for index in tenure_dist_pd_series.index:
    for value in range(tenure_dist_pd_series[index]):
        list_integers_to_replace.append(index)
```

```python
# Перемешаем значения
random.seed(rnd)
random.shuffle(list_integers_to_replace)
```

```python
# Заменим пропуски на значения созданного листа
df.tenure.loc[df.query('tenure.isna()').index] = list_integers_to_replace
```

```python
# Смотрим на итоговое распределение
df.tenure.value_counts()
```

```python
# Меняем тип данных в столбце
df['tenure'] = df['tenure'].astype('int')
```

```python
# Размер картинки
plt.gcf().set_size_inches(10, 6) 

# Строим график
tenure_hist = sns.histplot(df, x='tenure', kde=True, hue='exited', bins=21)

# Редактируем название осей и даем имя графику
_ = tenure_hist.set(title='Диаграмма времени удержания клиентов', xlabel="Время(в годах)", ylabel="Количество")
```

```python
# Смотрим остались ли пропуски
df.isna().sum()
```

Как мы видим характер распределения не поменялся, но мы избавились от пропусков.

### Масштабирование количественных признаков

Для того чтобы модели машинного обучения при обучении не присваивали большую важность признакам с большими значениями, произведем масштабирование с помощью `MinMaxScaler()`. 

На мой взгляд он более интерпритируемый чем `StandardScaler()` (отрицательные значения в столбце возраст могут сбить с толку)

```python
# Создаем экземпляр класса
scaler = MinMaxScaler()
# Определяем какие признаки будем масштабировать
col_for_scaling = ['credit_score', 'age', 'balance', 'estimated_salary']
```

```python
# Обучаем скейлер на обучающих данных
scaler.fit(X_train[col_for_scaling])

# Масштабируем
X_train[col_for_scaling] = scaler.transform(X_train[col_for_scaling])
X_valid[col_for_scaling] = scaler.transform(X_valid[col_for_scaling])
X_test[col_for_scaling] = scaler.transform(X_test[col_for_scaling])

# Смотрим результат
X_train.sample(2)
```

Для начала обучим модели без учета дисбаланса классов.

Создадим функцию построения графика матрицы ошибок:

```python
# На вход функции подаем матрицу ошибок 2*2 np.array() и название модели 
# Функция ничего не возвращает, просто подробно визуализирует матрицу ошибок
def make_conf_matrix(cf_matrix, model_name=None):    
    # Задаем параметры графика
    plt.rcParams["figure.figsize"] = [6.5, 6.11]
    plt.rcParams["figure.autolayout"] = True

    # Создаем привычные имена для  матрицы ошибок
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    
    # Количество в каждой группе
    group_counts = ['{0:0.0f}'.format(value) for value in
                    cf_matrix.flatten()]
    
    # Количество в каждой группе в процентах от всех данных
    group_percentages = ['{0:.2%}'.format(value) for value in
                        cf_matrix.flatten()/np.sum(cf_matrix)]

    # Создаем массив 2*2 с данными матрицы ошибок
    labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in
            zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)

    # Визуализируем матрицу ошибок
    sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues') \
       .set_title(f'Матрица ошибок для {model_name}')
```

### Логистическая регрессия

```python
def lr_func(X_train, X_valid, y_train, y_valid, class_weight=None, show_conf_matrix=True):

    # Задаем минимальное значение для целевой метрики
    best_lr_f1 = 0
    best_lr_model_f1 = None

    # Итерируемся по двум гиперпараметрам
    for c in tqdm(np.linspace(0.001, 10, 100), ncols = 100):
        for iter in range(10, 100, 10):
            # На каждой итерации создаем модель
            model_lr = LogisticRegression(C=c, max_iter=iter, n_jobs=-1, 
                                          class_weight=class_weight, random_state=rnd)
            # Обучаем
            model_lr.fit(X_train, y_train)
            # Делаем предсказание
            y_pred_lr = model_lr.predict(X_valid)
            # Считаем целевую метрику
            result_lr_f1 = f1_score(y_valid, y_pred_lr)
        
            # Сравниваем с лучшим результатом
            if result_lr_f1 > best_lr_f1:
                # Если больше, записываем в переменные значения полученные на текущей итерации
                best_lr_model_f1 = model_lr
                best_lr_f1 = result_lr_f1
        
    print(f'Лучший результат модели логистической регрессии по метрике f1: {best_lr_f1:.3f}')
    print('-' * 100)
    
    # Вычисляем ROC-AUC метрику для лучшей модели
    y_pred_proba = best_lr_model_f1.predict_proba(X_valid)
    roc_auc_lr = roc_auc_score(y_valid, y_pred_proba[:, 1])

    print(f'ROC-AUC метрика лучшей модели на основе логистической регрессии: {roc_auc_lr:.3f}')
    print('-' * 100)

    print(f'Параметры лучшей модели: \n{best_lr_model_f1.get_params()}')
    print('-' * 100)

    # Получаем матрицу ошибок
    cf_matrix_lr = confusion_matrix(y_valid, y_pred_lr)

    if show_conf_matrix == True:
        make_conf_matrix(cf_matrix_lr, 'LogisticRegression')

    return best_lr_model_f1, best_lr_f1, roc_auc_lr
```

```python
lr_unbalanced = lr_func(X_train, X_valid, y_train, y_valid)
```

* Как мы видим модель хорошо предсказывает нулевой класс, но лишь в одном случае из трех правильно предсказывает положительный класс.
* Вероятность того, что случайно выбранный экземпляр нулевого класса будет иметь меньше вероятность быть распознанным как позитивный класс, чем случайно выбранный позитивный класс (статистический смысл `roc_auc_score`) равняется `0.775`
* Максимальное значение метрики f1, которое удалось достичь перебором гиперпараметров: `0.349`, что далеко от необходимых `0.59`

### Дерево решений

```python
def dt_func(X_train, X_valid, y_train, y_valid, class_weight=None, show_conf_matrix=True):

    # Задаем минимальное значение для целевой метрики
    best_dt_f1 = 0
    best_dt_model_f1 = None

    # Итерируемся по четырем гиперпараметрам
    for depth in tqdm(range(1, 31), ncols = 100):
        for criterion in ['gini', 'entropy']:
            for min_split in range(2, 6):
                for min_leaf in range(1, 6):
                    model_dt = DecisionTreeClassifier(max_depth=depth, min_samples_split=min_split, min_samples_leaf=min_leaf,
                                                      class_weight=class_weight, random_state=rnd)
                                                      
                    # Обучаем
                    model_dt.fit(X_train, y_train)
                    # Делаем предсказание
                    y_pred_dt = model_dt.predict(X_valid)
                    # Считаем целевую метрику
                    result_dt_f1 = f1_score(y_valid, y_pred_dt)
        
                    # Сравниваем с лучшим результатом
                    if result_dt_f1 > best_dt_f1:
                        # Если больше записываем в переменные значения полученные на текущей итерации
                        best_dt_model_f1 = model_dt
                        best_dt_f1 = result_dt_f1
        
    print(f'Лучший результат модели дерева решений по метрике f1: {best_dt_f1:.3f}')
    print('-' * 100)

    # Вычисляем ROC-AUC метрику для лучшей модели
    y_pred_proba = best_dt_model_f1.predict_proba(X_valid)
    roc_auc_dt = roc_auc_score(y_valid, y_pred_proba[:, 1])

    print(f'ROC-AUC метрика лучшей модели на основе дерева решений: {roc_auc_dt:.3f}')
    print('-' * 100)

    print(f'Параметры лучшей модели: \n{best_dt_model_f1.get_params()}')
    print('-' * 100)

    # Получаем матрицу ошибок
    cf_matrix_dt = confusion_matrix(y_valid, y_pred_dt)

    if show_conf_matrix == True:
        make_conf_matrix(cf_matrix_dt, 'DecisionTreeClassifier')

    return best_dt_model_f1, best_dt_f1, roc_auc_dt
```

```python
dt_unbalanced = dt_func(X_train, X_valid, y_train, y_valid)
```

* Как мы видим модель чуть хуже предсказывает 0-ой класс нежели логистическая регрессия, но значительно более точно предсказывает класс 1
* Вероятность нашей модели различать классы значительно улучшилась по сравнению с логистической регрессией и стало равно `0.836`
* Максимальное значение метрики f1, которое удалось достичь перебором гиперпараметров: `0.599`, что больше необходимых `0.59`.

### Случайный лес

```python
def rf_func(X_train, X_valid, y_train, y_valid, class_weight=None, show_conf_matrix=True):
    
    # Задаем минимальное значение для целевой метрики
    best_rf_f1 = 0
    best_rf_model_f1 = None

    # Итерируемся по двум гиперпараметрам
    for estimators in tqdm(range(10, 151, 10), ncols = 100):
        for depth in range(5, 15):
            for criterion in ['gini', 'entropy']:
                model_rf = RandomForestClassifier(n_estimators=estimators, max_depth=depth, 
                                                  class_weight=class_weight, n_jobs=-1, 
                                                  random_state=rnd)
                # Обучаем
                model_rf.fit(X_train, y_train)
                # Делаем предсказание
                y_pred_rf = model_rf.predict(X_valid)
                # Считаем целевую метрику
                result_rf_f1 = f1_score(y_valid, y_pred_rf)
    
                # Сравниваем с лучшим результатом
                if result_rf_f1 > best_rf_f1:
                    # Если больше записываем в переменные значения полученные на текущей итерации
                    best_rf_model_f1 = model_rf
                    best_rf_f1 = result_rf_f1
        
    print(f'Лучший результат модели случайного леса по метрике f1: {best_rf_f1:.3f}')
    print('-' * 100)

    # Вычисляем ROC-AUC метрику для лучшей модели
    y_pred_proba = best_rf_model_f1.predict_proba(X_valid)
    roc_auc_rf = roc_auc_score(y_valid, y_pred_proba[:, 1])

    print(f'ROC-AUC метрика лучшей модели на основе случайного леса: {roc_auc_rf:.3f}')
    print('-' * 100)

    print(f'Параметры лучшей модели: \n{best_rf_model_f1.get_params()}')
    print('-' * 100)

    # Получаем матрицу ошибок
    cf_matrix_rf = confusion_matrix(y_valid, y_pred_rf)

    if show_conf_matrix == True:
        make_conf_matrix(cf_matrix_rf, 'RandomForest')

    return best_rf_model_f1, best_rf_f1, roc_auc_rf
```

```python
rf_unbalanced = rf_func(X_train, X_valid, y_train, y_valid)
```

* Как мы видим модель чуть лучше предсказывает 0-ой класс нежели дерево решений, но менее точно предсказывает класс 1
* Вероятность разделять наши классы у модели случайного дерева стала еще выше `0.861`
* Максимальное значение метрики f1, которое удалось достичь перебором гиперпараметров: `0.592`, что больше необходимых `0.59`.

### Градиентный бустинг

```python
def xgb_func(X_train, X_valid, y_train, y_valid, show_conf_matrix=True):

    # Задаем минимальное значение для целевой метрики
    best_xgb_f1 = 0
    best_xgb_model_f1 = None

    # Итерируемся по двум гиперпараметрам
    for estimators in tqdm(range(10, 151, 10), ncols=100):
        for depth in range(5, 15):
            for learning_rate in [0.001, 0.005, 0.07, 0.01, 0.05, 0.1]:
                model_xgb = XGBClassifier(n_estimators=estimators, max_depth=depth, learning_rate = learning_rate,
                                        n_jobs=-1,
                                        random_state=rnd)
                # Обучаем
                model_xgb.fit(X_train, y_train)
                # Делаем предсказание
                y_pred_xgb = model_xgb.predict(X_valid)
                # Считаем целевую метрику
                result_xgb_f1 = f1_score(y_valid, y_pred_xgb)

                # Сравниваем с лучшим результатом
                if result_xgb_f1 > best_xgb_f1:
                    # Если больше записываем в переменные значения полученные на текущей итерации
                    best_xgb_model_f1 = model_xgb
                    best_xgb_f1 = result_xgb_f1

    print(
        f'Лучший результат модели случайного леса по метрике f1: {best_xgb_f1:.3f}')
    print('-' * 100)

    # Вычисляем ROC-AUC метрику для лучшей модели
    y_pred_proba = best_xgb_model_f1.predict_proba(X_valid)
    roc_auc_xgb = roc_auc_score(y_valid, y_pred_proba[:, 1])

    print(
        f'ROC-AUC метрика лучшей модели на основе случайного леса: {roc_auc_xgb:.3f}')
    print('-' * 100)

    print(f'Параметры лучшей модели: \n{best_xgb_model_f1.get_params()}')
    print('-' * 100)

    # Получаем матрицу ошибок
    cf_matrix_xgb = confusion_matrix(y_valid, y_pred_xgb)

    if show_conf_matrix == True:
        make_conf_matrix(cf_matrix_xgb, 'XGB Regressor')

    return best_xgb_model_f1, best_xgb_f1, roc_auc_xgb
```

```python
xgb_unbalanced = xgb_func(X_train, X_valid, y_train, y_valid)
```

### Вывод:

* Максимального значения `f1-меры`, равного `0.599`, на валидационной выборке удалось достичь для модели решающего дерева. Также модель случайного дерева показала очень близкое значение `f1-меры`, равного `0.592`. При этом у модели случайного леса значение `ROC-AUC` выше чем у решающего дерева (`0,861` против `0,836`)

### Upsampling

Воспользуемся алгоритмом `SVMSMOTE` из библиотеки `imbalanced-learn`. 

В основе алгоритма `SMOTE` лежит идея генерации некоторого количества искусственных наблюдений, которые были бы «похожи» на наблюдения, имеющиеся в миноритарном классе, но при этом не дублируют их.

`SMOTE` использует алгоритм k-ближайших соседей для создания синтетических данных. Все начинается с выбора случайных данных из класса меньшинства, затем устанавливаются k ближайших соседей из данных. Затем создаются синтетические данные между случайными данными и случайно выбранным k-ближайшим соседом.

В нашем случае мы будем пользоваться алгоритмом `SVM-SMOTE`

Основные различия между `SVM-SMOTE` и `SMOTE` заключаются в том, что вместо использования k-ближайших соседей для выявления неправильной классификации этот метод будет включать алгоритм `SVM`(метод опорных векторов).

В `SVM-SMOTE` пограничная область аппроксимируется опорными векторами после обучения классификатора SVM на исходном обучающем наборе. Синтетические данные будут создаваться случайным образом вдоль линий, соединяющих вектор поддержки каждого класса меньшинства с рядом его ближайших соседей.

## Дополнительное решение с использованием кросс-валидации.

```python
# Разобъем нашу выборку на 2 части со следующим соотношением 80:20 (train:test)

# Разделим на обучающую и тестовую части. Для сохранения баланса классов используем stratify
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y, train_size=0.8, stratify=y, random_state=rnd)

# Проверим как разделилось
print(X_train_2.shape, y_train_2.shape)
print(X_test_2.shape, y_test_2.shape)
```

```python
# Создаем экземпляр класса
scaler_2 = MinMaxScaler()
# Определяем какие признаки будем масштабировать
col_for_scaling = ['credit_score', 'age', 'balance', 'estimated_salary']
# Обучаем скейлер на обучающих данных
scaler_2.fit(X_train_2[col_for_scaling])

# Масштабируем
X_train_2[col_for_scaling] = scaler_2.transform(X_train_2[col_for_scaling])
X_test_2[col_for_scaling] = scaler_2.transform(X_test_2[col_for_scaling])

# Смотрим результат
X_train_2.sample(2)
```

```python
# Создадим экземпляр класса
akn_2 = AllKNN(n_jobs=-1)
# Даунсэмплим
X_train_2_akn, y_train_2_akn = akn_2.fit_resample(X_train_2, y_train_2)
# Смотрим на количество полученных данных
print(X_train_2_akn.shape, y_train_2_akn.shape)
```

```python
# Создаем объект класса
model_2 = RandomForestClassifier(random_state=rnd)
# Параметры для перебора
params = {'n_estimators' : range(90, 131, 10), 'max_depth' : range (1, 21)}
# Создаем объект GridSearcCV с выбранными параметрами
grid_clf_rf = GridSearchCV(model_2, params, scoring = 'f1', cv=StratifiedKFold(n_splits=5), n_jobs=-1, verbose=10)
```

```python
# Ищем лучшие параметры на кросс-валидации
grid_clf_rf.fit(X_train_2_akn, y_train_2_akn)
```
